```python
from TwinNeuralNetwork import TNNR, ANNR
import matplotlib.pyplot as plt
```

```python
def_func = lambda a: lambda x: x**3 - a*x
```

```python
%%time

seed = 392
a = 80
n = 500

tnnr_64 = TNNR(def_func(a), n=n, seed=seed, 
               show_summary=False, show_loss_plot=False, show_rmse=False, mdl_wts_file='tnnr64n500')
annr_64 = ANNR(def_func(a), n=n, seed=seed, 
               show_summary=False, show_loss_plot=False, show_rmse=False, mdl_wts_file='annr64n500')
tnnr_128 = TNNR(def_func(a), n=n, seed=seed, neurons=128,
                show_summary=False, show_loss_plot=False, show_rmse=False, mdl_wts_file='tnnr128n500')
annr_128 = ANNR(def_func(a), n=n, seed=seed, neurons=128,
                show_summary=False, show_loss_plot=False, show_rmse=False, mdl_wts_file='annr128n500')

```

```python
plt.scatter(tnnr_64.cn_transformer.inverse_transform_x(tnnr_64.x_train_single), 
            tnnr_64.cn_transformer.inverse_transform_y(tnnr_64.y_train_single),
            label='Train data set', s=0.5, c='#99ffff')
plt.scatter(tnnr_64.cn_transformer.inverse_transform_x(tnnr_64.x_test_single), 
            tnnr_64.y_pred_test,
            label='TNNR $2\\times 64$ neurons', s=50)
plt.scatter(annr_64.cn_transformer.inverse_transform_x(annr_64.x_test_single), 
            annr_64.y_pred_test,
            label='ANNR $2\\times 64$ neurons', s=30)
plt.scatter(tnnr_128.cn_transformer.inverse_transform_x(tnnr_128.x_test_single), 
            tnnr_128.y_pred_test,
            label='TNNR $2\\times 128$ neurons', s=10)
plt.scatter(annr_128.cn_transformer.inverse_transform_x(annr_128.x_test_single), 
            annr_128.y_pred_test,
            label='ANNR $2\\times 128$ neurons', s=1, c='black')
plt.title('Function: $x^3 + 80x$')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.legend()
```

```python
plt.scatter(tnnr_64.cn_transformer.inverse_transform_x(tnnr_64.x_train_single), 
            tnnr_64.cn_transformer.inverse_transform_y(tnnr_64.y_train_single),
            label='Train data set', s=1, c='#00cccc')
plt.scatter(tnnr_64.cn_transformer.inverse_transform_x(tnnr_64.x_test_single), 
            tnnr_64.y_pred_test,
            label='TNNR $2\\times 64$ neurons', s=50)
plt.scatter(annr_64.cn_transformer.inverse_transform_x(annr_64.x_test_single), 
            annr_64.y_pred_test,
            label='ANNR $2\\times 64$ neurons', s=30)
plt.scatter(tnnr_128.cn_transformer.inverse_transform_x(tnnr_128.x_test_single), 
            tnnr_128.y_pred_test,
            label='TNNR $2\\times 128$ neurons', s=10)
plt.scatter(annr_128.cn_transformer.inverse_transform_x(annr_128.x_test_single), 
            annr_128.y_pred_test,
            label='ANNR $2\\times 128$ neurons', s=0.5, c='black')
plt.title('Function: $x^3 + 80x$')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.legend()
plt.xlim([-8, -7.35])
plt.ylim([130, 195])
```
